# Introdução ao Python

Computadores são conhecidos por terem a capacidade de fazer quase tudo. Cálculos, imagens, traduções, ou mesmo pensar. Essas máquinas, porém, não fazem nada independentemente. Tudo que um computador faz é seguir ordens. Tais instruções dadas aos computadores são chamados **algoritmos**.<br>
A palavra **algoritmo** é definida como uma sequência de passos a ser seguida a fim de atingir determinado objetivo. Por exemplo, uma receita de bolo é um algoritmo: temos passos a seguir com a finalidade de confeccionar o bolo. Para passarmos instruções às máquinas, então, criamos algoritmos a partir de **linguagens de programação**.

### O que é uma Linguagem de Programação?

Linguagens de programação são como métodos de comunicação entre os seres humanos e as máquinas (computadores), de modo que usamos tais linguagens para enviar comandos aos computadores. Computadores, no entanto, entendem apenas uma linguagem, a binária (constituída apenas de dois valores, 1 e 0). As linguagens de programação, por sua vez, são ferramentas utilizadas pelos seres humanos para programar máquinas de forma mais legível (sem ser apenas 1's e 0's) e palpável. Os códigos escritos em uma dada linguagem, então, são convertidos em linguagem de máquina após a compilação do código programado e esta, será exacutada pelo computador.

### Condições

Existem situações onde precisamos verificar ou testar se determinada condição é VERDADEIRA e, somente se ela for VERDADEIRA, vamos querer executar determinadas instruções. Veja um exemplo simples:\
Em um bar, o vendedor só deve vender bebidas **se** o cliente tiver mais de 18 anos. Caso o comprador seja menor de idade, a venda é proibida. Veremos agora como implementamos este tipo de situação em códigos de Python.

Usamos 'if', 'elif' e 'else' colocando uma condição ao lado destes comandos. Em Python, não se usa parênteses na condição. Ao final da condição, adiciona-se ':' para sinalizar o início do que será feito se a condição é verdadeira. Abaixo da linha do if, colocamos o que queremos que o computador faça, com uma identação (aquele espaço no início da linha) para indicar que está *dentro* da condição. Veja o exemplo do bar implementado:

````python
a = 16

if(a < 18):
  print("Proibida a venda.")
else:
  print("Venda permitida.")
````

````python
a = 18

if(a < 18):
  print("Proibida a venda.")
else:
  print("Venda permitida.")
````

Para criarmos as condições, usamos os seguintes sinais operadores:
1. == verifica se dois valores são **iguais**: 1 == 1 é verdadeiro;
2. != verifica se dois valores são **diferentes**: 1 != 2 é verdadeiro;
3. < verifica se o valor à esquerda é **menor** que o valor à direita: 1 < 2 é verdadeiro;
4. <= verifica se o valor à esquerda é **menor ou igual** que o valor à direita: 2 <= 2 é verdadeiro;
5. \> verifica se o valor à esquerda é **maior** que o valor à direita: 2 > 1 é verdadeiro;
6. \>= verifica se o valor à esquerda é **maior ou igual** que o valor à direita: 2 >= 2 é verdadeiro;

Caso precisarmos de mais de uma condição para executar uma tarefa, podemos usar a palavra chave ```and```.

````python
a = 1
b = 1
if(a < 2 and b < 2):
  print("Os dois são menores que 2!")
````

Caso tenhamos que executar o mesmo processo para duas condições diferentes, podemos usar a palavra chave ```or```.

````python
a = 1
b = 3
if(a < 2 or b < 2):
  print("Pelo menos um é menor que 2!")
````

### Loops

Existem situações em que queremos instruir o computador a executar várias instruções de forma repetida. Por exemplo: se você quiser que a palavra UFSC seja escrita na tela 100 vezes, você vai precisar programar 100 instruções print("UFSC")? Não, pois isso seria extremamente trabalhoso...\
Ao invés de reescrever as mesmas instruções dezenas, centenas ou milhares de vezes, podemos criar um **loop** para que essa tarefa seja efetuada tantas vezes quanto precisarmos de forma automática.

Para criarmos loops nos nossos códigos, podemos usar duas ferramentas específicas: ```for``` e o ```while```.\
Veremos abaixo, como executar um loop com cada uma das palavras-chave.

#### For

O ```for```, como o próprio nome indica, continua o loop <b>para</b> valores que estão dentro da região proposta pelo for. Para criarmos o loop, precisamos de uma condição que faz com que ele rode e **tenha um fim**.

Ainda não estudamos sobre listas, mas pense que é apenas um conjunto de itens colocados em uma mesma variável. Suponha que tenhamos uma lista de compras e queremos mostrar todos os produtos que precisamos comprar.

````python
lista = ['banana', 'maçã', 'leite']
````

Temos duas opções: escrever um ```print()``` para cada elemento ou usar um loop que mostra cada um automaticamente. No nosso caso, nossa lista possui apenas 3 itens. Não seria difícil escrever manualmente um print para cada um, não é mesmo? Mas imagine se tivéssemos centenas de itens na lista de compras. Teríamos que escrever manualmente a mesma linha de código para cada elemento centenas de vezes também. Assim, a opção dos loops é com certeza a melhor opção para repetições.\
Então, como ficaria um loop de ```for``` para este exemplo? Veja na célula a seguir.

````python
for elemento in lista:
    print(elemento)
````

Perceba que a palavra 'elemento', dentro do ```for```, é uma nova variável criada que vai carregar um elemento específico que aparece na lista. Conforme o loop é executado, o elemento muda, indo de um em um na lista. Em geral, um loop ```for``` vai seguir o formato:
```python
for elemento in lista_de_elementos
    ação a ser feita
    ...
```
Que literalmente significa fazer determinada ação **PARA** cada elemento na lista.

Mas e se quisermos executar uma mesma ação independente de uma lista múltiplas vezes?

##### Função **range()** 

Para estes casos que não temos uma lista inicial, só queremos repetir uma instrução, podemos criar um loop ```for``` usando a função ```range()```. A função ```range()``` nos retorna uma lista de valores, que vai de 0 até o valor colocado dentro da função. Então, se colocarmos o valor '10' dentro dos parênteses, a função retorna uma lista de números que vai de 0 até 9 (totalizando 10 valores).\
Você já deve estar percebendo que agora temos uma lista (criada pela função ```range()```), assim como no exemplo da lista de compras. A partir daqui, podemos criar o nosso loop da mesma forma que antes, usando uma variável auxiliar (geralmente usamos a letra ```i```) que vai passar por cada item da lista, assim como a variável ```elemento``` passava pela lista de compras no exemplo de antes.

````python
for i in range(10):
  print(i)
````

Assim, para fazer um loop se repetir X vezes, colocamos o tamanho X na função ```range(X)```.

Para o while, o loop continua <b>enquanto</b> tal condição é verdadeira. Novamente, é comum usarmos uma variável auxiliar ```i``` para criar a condição do nosso loop. Veja no exemplo abaixo: criamos a variável i e igualamos ela a zero. Nosso loop vai rodar **enquanto** i for menor que 10. Dentro do nosso loop, incrementamos o valor de i para que o loop tenha um fim (quando i chega a 10).

````python
i = 0
while i < 10:
    print(i)
    i = i+1
````

#### Ferramentas de Controle de Loop

Dentro de loops, podemos performar alguns passos a partir de palavras-chave do Python. Elas são: break e continue.

#### Break

A palavra-chave **break** é uma maneira de parar um loop quando determinada condição é satisfeita. Neste exemplo, criamos um loop que mostra os componentes de uma sequência, mas a partir do momento que encontra o primeiro valor acima de 5, o loop é impedido pelo **break**:

````python
sequence = [1, 2, 0, 4, 6, 5, 2, 1]

for i in range(10):
    if sequence[i] > 5:
        break
    print(sequence[i])
````

#### Continue

A palavra-chave **continue**, por sua vez, é usada para continuar o loop. Ela nada mais faz que pular para a próxima repetição do loop. No exemplo seguinte, perceba que o loop soma cada valor da dada sequência, exceto quando o valor dado é do tipo *None*. O **continue** é útil neste tipo de caso em que dados problemas podem ser ignorados sem a interrupção do loop:

````python
sequence = [1, 2, None, 4, None, 5]
total = 0
for value in sequence:
  if value is None:
    print('Valor None! Total: %d' %(total))
    continue
  total = total + value
  print('Total: %d' %(total))
````

# Estruturas de Dados e Seus Métodos

Estruturas de dados são formas de armazenamento de grupos de dados. Cada estrutura tem sua função e tem suas próprias características e métodos. Python possui diversas estruturas de dados disponíveis. Veremos algumas das mais comuns e úteis.

## Tuplas

Uma tupla é uma sequência de objetos de tamanho fixo:

````python
tup = 4, 5, 6

tup
````

Para definir tuplas mais complicadas, é recomendado utilizar valores entre parênteses. Isso é útil para colocar estruturas mais complexas que apenas números em uma tuplas, como por exemplo uma tupla de tuplas:

````python
tupla_inception = (4, 5, 6), (7, 8)

tupla_inception
````

Podemos ainda converter uma dada sequência em uma tupla ao invocarmos a função ```tuple()```:

````python
print(tuple([4,0,2]))

print(tuple("texto"))
````

Podemos acessar os elementos colocando entre colchetes [] o índice do elemento buscado.

````python
minhatupla = 0, 3, 4

minhatupla[1]
````

Tuplas podem ter multiplos elementos de diferentes classes:

````python
tupla = tuple(['texto', [1, 4], True, 5.75])

print(tupla)

tupla[1].append(5)

print(tupla)
````

Podemos ainda juntar tuplas com o operador +:

````python
(41, None, 'aaa') + (6,0) + ('bbbb',)
````

Multiplicar tuplas por números inteiros faz com que a tupla se concatene consigo mesma o número de vezes multiplicado:

````python
('sou', 'lindo') * 4
````

Se igualarmos uma expressão de variáveis cuja estrutura condiga com a de uma determinada tupla, os valores serão desempacotados de forma organizada:

## Listas

Listas são como tuplas, mas em que seus elementos podem ser modificados. Podemos defini-las colocando dados entre colchetes ou usando a função ```list()```

````python
lista_de_compras = ['a', 'b', 'c']
lista_de_compras
````

Podemos alterar um valor da lista chamando o valor pelo seu índice:

````python
tupla_de_compras = ('banana', 'maçã', 'berga')
lista_de_compras = list(tupla_de_compras)
print(lista_de_compras)

lista_de_compras[0] = 'banana prata'
print(lista_de_compras)
````

````python
conta = list(range(10))
conta
````

### Adicionando e Removendo Elementos

Podemos adicionar um dado ao final de uma lista usando a função ```append()```:

````python
lista_de_compras.append('pão')
lista_de_compras
````

Para adicionar um dado em uma localização específica da lista, usa-se a função ```insert()```, colocando o índice onde queremos colocar em frente ao valor:

````python
lista_de_compras.insert(1, 'manteiga')
lista_de_compras
````

A operação inversa seria a função ```pop()```, que remove e retorna o elemento no índice indicado:

````python
lista_de_comprados = []
lista_de_comprados.append(lista_de_compras.pop(2))

print(lista_de_compras)
print(lista_de_comprados)
````

Para remover um dado usando seu valor, e não seu índice, usamos a função ```remove()```:

````python
lista_de_compras.remove("pão")

print(lista_de_compras)
````

### ```zip()```

A função ```zip()``` pareia elementos de listas, tuplas ou outras sequências para criar uma lista de tuplas:

````python
seq1 = ['foo', 'bar', 'baz']
seq2 = ['one', 'two', 'three']
zipped = zip(seq1, seq2)
list(zipped)
````

````python
seq3 = [False, True]
list(zip(seq1, seq2, seq3))
````

````python
for i, (a, b) in enumerate(zip(seq1, seq2)):
    print('{0}: {1}, {2}'.format(i, a, b))
````

````python
pitchers = [('Nolan', 'Ryan'), ('Roger', 'Clemens'),
            ('Schilling', 'Curt')]
first_names, last_names = zip(*pitchers)
first_names
last_names
````

## Dicionários

Dict é muitas vezes considerada a estrutura mais importante disponível em Python. É possível nomear dados dentro de uma estrutura. Podemos definir um dict colocando entre chaves os valores que queremos:

````python
dict_vazio = {}
dict_vazio
````

````python
dict_1 = {'a' : '1', 'b' : [2, 3, 5, 1]}
dict_1
````

Podemos acessar, inserir ou alterar elementos usando a mesma sintaxe para acessar elementos de uma lista ou tupla:

````python
dict_1[7] = 'alguma coisa'
dict_1
````

````python
dict_1['b']
````

Podemos checar se uma dict contém uma chave específica usando a palavra-chave <b>in</b>:

````python
'b' in dict_1
````

Para deletar valores, usamos ou a palavra-chave <b>del</b> ou o método ```pop``` demonstrada anteriormente:

````python
dict_1[10] = "sacrifício"
dict_1['aaa'] = "sacrifício2"

print(dict_1)

del dict_1[10]
dict_1.pop('aaa')

print(dict_1)
````

Podemos listar os índices/nomes dos elementos usando a função ```keys()``` e os valores com a função ```values()```:

````python
list(dict_1.keys())
````

````python
list(dict_1.values())
````

Podemos ainda combinar ou alterar uma dict em outra usando o método ```update()```:

````python
dict_1.update({'b' : 'aaaa', 'c' : 12})
dict_1
````

### Criando Dicts a partir de Sequências

Podemos mapear sequências usando uma mistura da função ```dict()``` com ```zip()```:

````python
mapear_sequencia = dict(zip(range(5), reversed(range(5))))
mapear_sequencia
````

## Sets

Sets são um conjunto ordenado de elementos <b>únicos</b>. Seriam como dicts que tem apenas chaves de elementos, sem valores. Podemos criar um set a partir da função ```set()``` ou colocando valores entre chaves:

````python
set([2,1,2,3,4,5,3,3,3,2,2,2,1])
````

````python
{2,2,2,1,1,3,3,4,2}
````

### Operações de Sets

Podemos unir dois sets com a função ```union()```:

````python
a = {1, 2,3,4,1,1,1,2,1}
b = {3,4,3,2,5,6}

a.union(b)
````

Para encontrarmos intersecções, usamos a função ```intersection()``` ou o sinal <b>&</b>:

````python
a.intersection(b)
````

````python
a & b
````

O método ```copy()``` faz uma cópia de um dado set:

````python
c = a.copy()
c
````

Podemos checar se um determinado set é subset de outro com o método ```issubset()```:

````python
a_set = {5,3,4,2,1}
a_set
````

````python
{1, 2, 3}.issubset(a_set)
````

Podemos fazer o caminho contrário e ver se um set é *superset* de outro:

````python
a_set.issuperset({1, 2, 3})
````
