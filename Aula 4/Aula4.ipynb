Podemos adicionar um t√≠tulo pro nosso gr√°fico com o m√©todo `title()`:

### Nomea√ß√£o de Eixos

Podemos colocar nomes nos eixos $x$ e $y$ com os m√©todos `xlabel()` e `ylabel()`, respectivamente:

```python
plt.plot([1,2,3], [2,4,6])

plt.title('nosso gr√°fico!!!')

plt.xlabel('eixo X')

plt.ylabel('eixo Y')

plt.show()
```

Vamos tentar plotar dois gr√°ficos juntos? O jeito mais f√°cil de fazer isso √© chamando o `.plot` duas vezes.

```python
x1 = [1,2,3,4,5]
y1 = [2,4,6,8,10]

x2 = [0,2,4,8,16]
y2 = [4,4,4,4,4]

plt.plot(x1,y1)
plt.plot(x2,y2)

plt.title('gr√°fico DUPLO')

plt.xlabel('eixo X')

plt.ylabel('eixo Y')

plt.show()
```

Beleza! Plotamos dois gr√°ficos juntos. Mas t√° meio feia a nossa imagem de retorno, n√£o acha? Vamos tentar mudar isso.\
Usando o m√©todo `.figure()`, podemos alterar as configura√ß√µes da imagem do gr√°fico que o matplot nos retorna. Dentre os par√¢metros que podemos colocar dentro do `.figure()`, temos o *figsize* e o *dpi*. O *figsize* √© usado pra escolhermos as propor√ß√µes da nossa imagem. Colocando $(3,4)$ para o *figsize*, por exemplo, o gr√°fico vai retornar com uma propor√ß√£o 3x4.\
J√° o *dpi* serve para realmente mudarmos o tamanho do nosso gr√°fico. *dpi* indica o n√∫mero de pontos por polegada (*dots per inch*). Ent√£o, quanto maior o n√∫mero que colocarmos, maior ser√° o gr√°fico.\
Enfim, bora deixar nosso gr√°fico maior! Vamos colocar a propor√ß√£o 5x3 e o *dpi* em 150:

```python
plt.figure(figsize=(5,3), dpi=150)

plt.plot(x1,y1)
plt.plot(x2,y2)

plt.title('gr√°fico DUPLO')

plt.xlabel('eixo X')

plt.ylabel('eixo Y')

plt.show()
```

## Tipos de Gr√°ficos

Como todos devem saber, existem diversos tipos de gr√°ficos, gr√°ficos de setores (popularmente conhecido como gr√°fico de pizzas), linha, barras, etc. Vamos tentar explorar um pouco dos mais usados.

### Gr√°ficos de Pizza

Gr√°ficos do tipo pizza s√£o um dos tipos de gr√°ficos mais f√°ceis e simples de se entender.

Vamos tentar fazer um gr√°fico do tipo Pizza que mostra a porcentagem da popula√ß√£o de 2022 separando cada continente.

Podemos criar um gr√°fico de pizza com o m√©todo `.pie()`. No m√©todo, podemos introduzir uma lista com os valores que queremos. Para colocarmos as legendas, adicionamos o par√¢metro `labels` e colocamos a lista de nomes que representam cada n√∫mero. Temos que colocar os nomes na mesma ordem dos valores!!! 

Por fim, podemos colocar o par√¢metro `autopct` pra que possamos escolher como nossa porcentagem aparece. Nesse caso, vou usar `'%.2f%%'`. O primeiro % √© para indicar que os pr√≥ximos caracteres ser√£o um valor. O .2f pede que o n√∫mero seja representado como um float com 2 d√≠gitos depois da v√≠rgula. Por fim, o %% serve para mostrar o sinal de porcentagem depois do n√∫mero. Sim, precisamos colocar o dois sinais (%%) ao inv√©s de 1, sen√£o o Python interpreta que deveria ter algum n√∫mero ali.

Come√ßamos criando nossas listas:

```python
asia_pop=wrld_pop.loc[wrld_pop['Continent'] == 'Asia', ['2022 Population']].sum()
sa_pop=wrld_pop.loc[wrld_pop['Continent'] == 'South America', ['2022 Population']].sum()
africa_pop=wrld_pop.loc[wrld_pop['Continent'] == 'Africa', ['2022 Population']].sum()
na_pop=wrld_pop.loc[wrld_pop['Continent'] == 'North America', ['2022 Population']].sum()
oce_pop=wrld_pop.loc[wrld_pop['Continent'] == 'Oceania', ['2022 Population']].sum()
eu_pop=wrld_pop.loc[wrld_pop['Continent'] == 'Europe', ['2022 Population']].sum()

continent_pops=[asia_pop['2022 Population'],sa_pop['2022 Population'],africa_pop['2022 Population'],na_pop['2022 Population'],oce_pop['2022 Population'], eu_pop['2022 Population']]
labels=['Asia', 'South America', 'Africa', 'North America', 'Oceania', 'Europe']
```

```python
plt.figure(figsize=(5,3), dpi=150)
plt.pie(continent_pops, labels=labels, autopct='%.2f%%')
plt.show()
```

Ba, mas deu um trabalh√£o montar todas aquelas listas... ser√° que n√£o tem um jeito mais f√°cil de fazer esse gr√°fico? ü§®

Relaxa, porque tem sim! üôè\
A pr√≥pria biblioteca Pandas tem um m√©todo plot que nos deixa plotar gr√°ficos de forma r√°pida com nossos datasets. Pra fazer o mesmo gr√°fico que criamos ali em cima, podemos agrupar nosso dataset original ``wrld_pop`` pelos continentes usando o m√©todo `groupby()`. Como queremos somar as popula√ß√µes, botamos o m√©todo `sum()`.

```python
continentes = wrld_pop.groupby('Continent').sum()
continentes
```

Veja que agora temos um novo dataset que agrupou todos os pa√≠ses pelos seus continentes e somou todos os valores de dentro das colunas. Alguns dados somados ali dentro n√£o fazem sentido, ent√£o podemos retir√°-los.

```python
continentes.drop(['Rank', 'Density (per km¬≤)', 'Growth Rate'], axis=1, inplace=True)
continentes
```

Beleza! Agora podemos montar nosso gr√°fico!

```python
continentes.plot(kind='pie', y='World Population Percentage', autopct='%.2f%%', legend=False)

plt.show()
```
