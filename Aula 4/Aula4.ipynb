## Plots Iniciais

Enfim, bora plotar uns gr√°ficos!

Vamos come√ßar com um exemplo simples: plotar uma fun√ß√£o $f(x)=x$. Para fazer qualquer gr√°fico, usamos o m√©todo `.plot` da nossa biblioteca.

```python
plt.plot([1,2,3],[2,4,6])

plt.show()
```

Vamos entender o que aconteceu ali em cima.\
Dentro da nossa fun√ß√£o `plot`, colocamos duas listas, a `[1,2,3]` e a `[2,4,6]`. A primeira lista que colocamos, se refere aos valores de $x$ no gr√°fico, enquanto a segunda representa os valores de $y$. Temos como resultado, um conjunto de pontos: (1,2), (2,4) e (3,6). A pr√≥pria biblioteca vai se encarregar de ligar todos os pontos que criamos, fazendo a linha que mostramos acima.

Antes de tentarmos fazer gr√°ficos muito complexos, vamos aprender o b√°sico pra alterarmos a apar√™ncia dos nossos gr√°ficos.

### T√≠tulo

Podemos adicionar um t√≠tulo pro nosso gr√°fico com o m√©todo `title()`:

```python
plt.plot([1,2,3],[2,4,6])

plt.title('nosso gr√°fico')

plt.show()
```

### Eixos nomeados

Podemos colocar nomes nos eixos $x$ e $y$ com os m√©todos `xlabel()` e `ylabel()`, respectivamente:

```python
plt.plot([1,2,3], [2,4,6])

plt.title('nosso gr√°fico!!!')

plt.xlabel('eixo X')

plt.ylabel('eixo Y')

plt.show()
```
