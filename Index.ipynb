## Resumo

Como podemos perceber nos dias de hoje, existe uma grande quantidade de dados sendo gerados, tanto em instituies pblicas como privadas. Saber manipular esses dados  uma importante habilidade que pode ajudar a tomar as melhores decises nessas instituies.
