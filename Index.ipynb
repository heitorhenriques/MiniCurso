## Resumo

Como podemos perceber nos dias de hoje, existe uma grande quantidade de dados sendo gerados, tanto em instituições públicas como privadas. Saber manipular esses dados é uma importante habilidade que pode ajudar a tomar as melhores decisões nessas instituições.

Por isso o foco desse curso é introduzir, de forma bem simples, à análise de dados com Python. Aqui você irá aprender um pouco de como carregar, limpar e visualizar esses dados.

Este minicurso foi desenvolvido por um projeto de extensão que é o GEPy, fazendo parte da Liga Academica de Inteligência Artificial (LIA), os integrantes que fizeram parte do desenvolvimento desse minicurso esse ano foram, Bernardo Pandolfi Costa, Eduardo Cunha Cabral, Heitor Henrique da Silva e a professora Andréa Sabedra Bordin.
